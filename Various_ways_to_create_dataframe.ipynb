```python
#From a list of list
import pandas as pd
data=[['Alice',25,'Accountant','12,000'],['John','28','Engineer','30,000'],['Julie','26','Teacher','25,000'],['Mark',30,'Doctor','29,000']]
df=pd.DataFrame(data, columns=['Name','Age','Profession','Salary'])
df
```

```python
#From a dictionary
data={'Name':['Alice','John','Julie','Mark'],'Age':[25,28,26,30],'Profession':['Account','Engineer','Teacher','Doctor'],'Salary':['12,000','30,000','25,000','29,000']}
df=pd.DataFrame(data)
df
```

```python
#From a numpy array
import numpy as np
data=np.array([['Alice',25,'Accountant','12,000'],['John','28','Engineer','30,000'],['Julie','26','Teacher','25,000'],['Mark',30,'Doctor','29,000']])
df=pd.DataFrame(data)
df
```

```python
#From JSON
data='{"Name":["Alice","John","Julie","Mark"],"Age":[25,28,26,30],"Profession":["Account","Engineer","Teacher","Doctor"],"Salary":["12,000","30,000","25,000","29,000"]}'
df = pd.read_json(data)
df
```
